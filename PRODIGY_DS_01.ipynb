```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset (ages)
ages = np.random.randint(18, 80, size=100) # generate 100 random ages between 18 and 80

# Create a histogram to visualize the distribution of ages
plt.figure(figsize=(8, 6))
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Ages in population")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.show()
```
